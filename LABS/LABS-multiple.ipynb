```
import numpy as np

def binary_to_signed(binary_sequence):
    """
    Convert binary sequence (0, 1) to signed sequence (-1, 1).
    s_i = 2 * x_i - 1
    """
    return 2 * np.array(binary_sequence) - 1

def calculate_autocorrelation(signed_sequence):
    """
    Calculate the autocorrelation sum for a given signed sequence.
    """
    n = len(signed_sequence)
    autocorrelation_sum = 0

    # Iterate over all possible shifts k
    for k in range(1, n):  # k starts from 1 (non-zero shifts)
        autocorrelation_k = 0
        for i in range(n - k):  # Compute autocorrelation for shift k
            autocorrelation_k += signed_sequence[i] * signed_sequence[i + k]
        autocorrelation_sum += autocorrelation_k ** 2  # Square and accumulate

    return autocorrelation_sum

def labs_objective(binary_sequence):
    """
    Compute the LABS objective value for a binary sequence.
    """
    n = len(binary_sequence)
    signed_sequence = binary_to_signed(binary_sequence)
    autocorrelation_sum = calculate_autocorrelation(signed_sequence)

    # LABS formula
    labs_score = n ** 2 / (2 * autocorrelation_sum)
    return labs_score

# Example usage
binary_sequence = [0, 0, 0, 1]
print("Binary sequence:", binary_sequence)
labs_score = labs_objective(binary_sequence)
print("LABS score:", labs_score)


```
